# Золотое сечение

```python
def func_3(x):
    return x**2 + np.abs(np.sin(x))

t = np.arange(-1/2, 5/2, 0.0001)
y = func_3(t)

def golden_ratio(func, a, b, eps = 0.001):
    x2 = a + 0.61803 * (b - a)
    x1 = a + b - x2
    f1 = func(x1)
    f2 = func(x2)
    plt.plot(t, y, '--g')
    plt.plot(x1, f1, 'or', x2, f2, 'ob')
    while(b - a > 2 * eps):
        if(f1 < f2):
            b = x2
            x2 = x1
            f2 = f1
            x1 = a + b - x2
            f1 = func(x1)
            print(a, b)
        else:
            a = x1
            x1 = x2
            f1 = f2
            x2 = a + b - x1
            f2 = func(x2)
            print(a, b)
        #plt.plot(x1, f1, 'or', x2, f2, 'ob')
    x = (a + b) / 2
    plt.show()
    return x
```

# Метод перебора

```python
def iterating_method(func, a, b, n = 1000):
    length_of_iter = (b - a) / (n + 1)
    min_x = a
    x_i = min_x
    for i in range(n):
        x_i += length_of_iter
        if (func(min_x) > func(x_i)):
            min_x = x_i
    return min_x
```

```python
t = np.arange(-1, 1, 0.001)
y1 = func_1(t)
y2 = func_2(t)

point1 = iterating_method(func_1, -1., 1.)
print(f'x1: {point1:.10f}, \t\ty1: {func_1(point1)}')

plt.subplot(2, 1, 1)
plt.plot(t, y1, '--b', point1, func_1(point1), 'or')

point2 = iterating_method(func_2, -0.6, 0.8)
print(f'x2: {point2:.10f}, \t\ty2: {func_1(point2)}')
plt.subplot(2, 1, 2)
plt.plot(t, y2, '--b', point2, func_2(point2), 'or')
plt.xlim(-0.6, 0.6)
plt.ylim(-0.3, 1)

plt.show()
```
